```python
import os 
```

```python
def set_working_directory(folder_path):
    os.chdir(folder_path)
    new_directory = os.getcwd()
    print("Patrick's Current working directory is set to:", new_directory)

# Set the working directory back to the original folder
folder_path = r"C:\Users\patme\OneDrive\Desktop\Data_Science"
set_working_directory(folder_path)
```

```python
# Up to down approach
def top_down_walk():
    for dirpath, dirnames, files in os.walk('Python/'):
        print("Directory:", dirpath)
        print("Include these directories")
        for dirname in dirnames:
            print(dirname)
        print("Include these files")
        for filename in files:
            print(filename)
        print()

if __name__ == "__main__":
    top_down_walk()
```

```python
# Bottom to up approach

def botton_up_walk():
    for dirpath, dirnames, files in os.walk('Python/', topdown=False):
        print("Directory:", dirpath)
        print("Include these directories")
        for dirname in dirnames:
            print(dirname)
        print("Include these files")
        for filename in files:
            print(filename)
        print()

if __name__ == "__main__":
    botton_up_walk()
```
